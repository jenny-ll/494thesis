# create dictionary to map HS code products to their respective EXIO code

```python
bridge = pd.read_excel('../Data/3_Trade/hs96_exio20_bridge.xlsx')

bridge.columns = ['hs1996',

                 'exiobase_20',

                 'exiobase_20_2']

bridge_tuples = list(bridge[['hs1996', 'exiobase_20']].itertuples(index=False, name=None))
bridge_tuples_2 = list(bridge[['hs1996', 'exiobase_20_2']].itertuples(index=False, name=None))
bridge_tuples_no_nan = [t for t in bridge_tuples_2 if not any(isinstance(n, float) and math.isnan(n) for n in t)]

full_bridge = bridge_tuples_no_nan + bridge_tuples
full_bridge
```

```python
full_bridge_dict = dict(full_bridge)
full_bridge_dict
```

```python
code_df = pd.DataFrame(full_bridge_dict.items(), columns=['HS_code', 'exio_code'])
```

```python
merged_df = pd.merge(df, code_df, on="HS_code")
```

```python
merged_df.head()
```

# Load dataset from OrganizeTradeData

```python
em_df = pd.read_csv('../Data/3_Trade/product_emissions_with_exiocode.csv')
```

```python
em_df.head()
```

### Merge datasets, based on exporter (since carbon emissions is dependent on the exporter)

```python
final_merged_df = pd.merge(merged_df, em_df,  how='left', left_on=['year','exporter','exio_code'], right_on = ['year','country_number','exio_code'])
```

```python
final_merged_df.head()
```

```python
final_merged_df = final_merged_df.drop('country_number', axis=1)
sorted_merged_df = final_merged_df.sort_values('year')
sorted_merged_df = sorted_merged_df.reset_index(drop=True)
sorted_merged_df['carbon_times_quantity'] = sorted_merged_df['quantity_tons']*sorted_merged_df['total_co2']
sorted_merged_df
```

```python
merged_df_noempty = sorted_merged_df.dropna()
```

```python
merged_df_noempty
```

```python
merged_df_noempty['total_co2'].describe()
```

### For each year and importer and HS_code, find quantity sum imported and weighted average of carbon emissions of exporters for that year and product

```python
good_df = merged_df_noempty.groupby(
    ['year','importer','product_name','HS_code']
).agg(
    quantity_sum = ('quantity_tons','sum'),
    mean_emissions  = ('carbon_times_quantity','mean'),
).reset_index()

good_df
```

```python
good_df.to_csv('../Data/3_Trade/improvedmapping_organized_trade.csv',index=False)
```
